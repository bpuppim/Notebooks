##### <font size=2 color = '#ff79c6'>Cross Validation</font>

```python
def cross_val_f1(X, y, model, model_name):
    
    kf = KFold(n_splits=30) #fazer com 30
    metric = 'f1'

    scores = cross_val_score(model, X, y, cv=kf, scoring=metric)
    
    return pd.DataFrame( {'Model Name': model_name,
                          'F1 SCORE  CV': np.round( scores.mean(), 5 ).astype( str ) + ' +/- ' + np.round( scores.std(), 2 ).astype( str ) }, index=[0] )
```

##### <font size=2 color = '#ff79c6'>Outliers</font>

```python
def remove_outliers_tukey(data,lower,upper):
    q1, q3 = np.percentile(data, [lower, upper])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    return data[(data >= lower_bound) & (data <= upper_bound)]
```

##### <font size=2 color = '#ff79c6'>Gráfico P-Value</font>

```python
def TestStatisticPlot( pvalue, alpha=0.05 ):
    # Gerando a distribuição normal padrão com média 0 e desvio padrão 1
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    # Calculando os limites do intervalo de confiança
    z = norm.ppf(1-alpha/2)
    x1 = np.linspace(-4, -z, 100)
    y1 = norm.pdf(x1, 0, 1)
    x2 = np.linspace(z, 4, 100)
    y2 = norm.pdf(x2, 0, 1)

    # Plotando uma linha vertical para o valor de p
    if stats.norm.ppf(pvalue) < -4:
        z_ = -4
    elif stats.norm.ppf(pvalue) > 4:
        z_ = 4
    else:
        z_ = stats.norm.ppf(pvalue)

    # Criando o gráfico
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.axvline(z_, color='#ff79c6')
    ax.plot(x, y, color='#6272a4', linewidth=2)
    ax.fill_between(x1, y1, color='#bd93f9', alpha=0.5)
    ax.fill_between(x2, y2, color='#bd93f9', alpha=0.5)
    ax.set_title('Distribuição normal padrão com área pintada referente a alpha={}'.format(alpha))

    # Adicionando títulos e rótulos
    plt.title('Distribuição Normal (Gaussiana) p-value e alpha')
    plt.xlabel('Estatística Z')
    plt.ylabel('Densidade de Probabilidade')
    ax.text( (z_ + 0.1) , 0.20, 'P-VALUE', fontsize=12, color='#ff79c6')
    ax.text( (-4 -z)/2 , 0.05, 'ALT HYPOTHESIS',   fontsize=14, color='#bd93f9', ha='center')
    ax.text( (-4 -z)/2 , 0.025, alpha/2,   fontsize=14, color='#bd93f9', ha='center')
    ax.text( (+4 +z)/2 , 0.05, 'ALT HYPOTHESIS',   fontsize=14, color='#bd93f9', ha='center')
    ax.text( (+4 +z)/2 , 0.025, alpha/2,   fontsize=14, color='#bd93f9', ha='center')
    ax.text( 0 , 0.02, 'NULL HYPOTHESIS',  fontsize=14, color='#6272a4', ha='center')

    plt.show()
    return
```

##### <font size=2 color = '#ff79c6'>Features Transformation </font>

```python
def TransformBinary (df):
    #===================#
    # ORIGINALS BINARY  #
    #===================#
    enc_binary = OneHotEncoder( drop='if_binary', 
                                sparse=False )
    df[['cliente']]            = enc_binary.fit_transform(df[['cliente']]           ).astype(int)
    df[['agencia_de_turismo']] = enc_binary.fit_transform(df[['agencia_de_turismo']]).astype(int)
    df[['por_empresa']]        = enc_binary.fit_transform(df[['por_empresa']]       ).astype(int)
    df[['estacionamento']]     = enc_binary.fit_transform(df[['estacionamento']]    ).astype(int)
    df[['qnt_estrelas']]       = enc_binary.fit_transform(df[['qnt_estrelas']]      ).astype(int)

    with open('targt_encoder1.pickle', 'rb') as f:
        freq_encoder1 = pickle.load(f)
    with open('targt_encoder2.pickle', 'rb') as f:
        freq_encoder2 = pickle.load(f)
    
    df['tipo_quarto']     = targt_encoder1.transform(df['tipo_quarto']    )
    df['forma_dereserva'] = targt_encoder2.transform(df['forma_dereserva'])

    #===================#
    #    ALIMENTACAO    #
    #===================#
    ohe_alimentacao = pd.get_dummies(df['alimentacao'])
    # Rename
    ohe_alimentacao = ohe_alimentacao.rename(columns={'Café da manha' : 'cafe' ,
                                                      'Sem refeicao' : 'sem_refeicao' , 
                                                      'Café da manha e jantar' : 'cafe_janta' ,
                                                      'Café da manha, almoco e jantar': 'cafe_almoco_janta'})    
    # concat new dummies
    df = pd.concat([df.drop('alimentacao', axis=1), ohe_alimentacao], axis=1)


    #===================#
    #        OBS        #
    #===================#
    df['obs'] = df['obs'].apply(lambda x: 1 if x == 'Nenhuma' else 0)


    return df
```

```python
def TransformOutliers(df):

    df['numero_pessoas']   = remove_outliers_tukey(df['numero_pessoas'],0,95)
    df['duracao_reservas'] = remove_outliers_tukey(df['duracao_reservas'], 0,95)
    df['antecedencia']     = remove_outliers_tukey(df['antecedencia'],0,95)

    df['numero_pessoas']   = df['numero_pessoas'  ].fillna(df['numero_pessoas'].max())
    df['duracao_reservas'] = df['duracao_reservas'].fillna(df['duracao_reservas'].max())
    df['antecedencia']     = df['antecedencia'    ].fillna(df['antecedencia'].max())

    
    return df
```

##### <font size=2 color = '#ff79c6'>Features Engineering </font>

```python
def CreateLockdownFeature (df):    
    
    # create bins
    bins = list( np.arange( 0 ,121001 ,1000 ) )
    df['bin'] = pd.cut( df['id'], bins=bins ).apply(lambda x: float(x.right)).astype(int)

    # Inserir valor da feature nova
    df_lockdown = pd.read_pickle('df_lockdown.pickle')
    df = pd.merge( df, df_lockdown, on='bin')
    df = df.drop(columns='bin', axis=1)
    
    return df
```

##### <font color = '#ff79c6' size=3>Intervalo de Confiança da diff média

```python
# Amostras das duas populações independentes
d1 = aux[aux['reserva_cancelada']==0]['duracao_reservas']
d2 = aux[aux['reserva_cancelada']==1]['duracao_reservas']

# médias
m1 = np.mean(d1)
m2 = np.mean(d2)

# std
std1 = np.std(d1, ddof=1)
std2 = np.std(d2, ddof=1)

# tamanho da amostra
n1 = len(d1)
n2 = len(d2)

# confiança
confianca = 0.95

# Valor crítico de z
z = norm.ppf((1 + confianca) / 2)

# Cálculo do intervalo de confiança z para a diferença da média das duas populações independentes
ic_inferior = (m1 - m2) - z * np.sqrt(std1**2/n1 + std2**2/n2)
ic_superior = (m1 - m2) + z * np.sqrt(std1**2/n1 + std2**2/n2)

print(f'Intervalo de confiança {confianca*100}% para a diferença da média das duas populações independentes: ({ic_inferior:.2f}, {ic_superior:.2f})')
```

```python
aux = df2[['id','reserva_cancelada']]

# create a bin list
bins = list( np.arange( 0 ,121000 ,1000 ) )

# apply cut func. on bin's list 
aux['BINS'] = pd.cut( aux['id'], bins=bins )
aux_bins = aux.groupby(by='BINS').sum().reset_index()
aux_bins['BINS'] = aux_bins['BINS'].apply(lambda x: float(x.right))

aux2_bins = aux.groupby(by='BINS').count().reset_index()
aux2_bins['BINS'] = aux2_bins['BINS'].apply(lambda x: float(x.right))

# Calcular a média móvel
aux_bins ['media_movel'] = aux_bins ['reserva_cancelada'].rolling(window=3).mean()
aux2_bins['media_movel'] = aux2_bins['reserva_cancelada'].rolling(window=3).mean()


# Criar Gráfico
plt.plot( aux_bins['BINS'], aux_bins['reserva_cancelada'], color='#bd93f9', label='Cancelamentos')
plt.plot(aux2_bins['BINS'],aux2_bins['reserva_cancelada'], color='#6272a4', label='Total Reservas')
plt.plot(aux_bins['BINS'] ,aux_bins ['media_movel'], color='#ff79c6', linestyle='dashed', alpha=0.5)
plt.plot(aux2_bins['BINS'],aux2_bins['media_movel'], color='#8be9fd', linestyle='dashed', alpha=0.5)


# Adicionar rótulos e título
plt.xlabel('Intervalo de id')
plt.ylabel('Quantidade de Reservas')
plt.title('Comportamento das Reservas no "pseudo-tempo"')

plt.legend()
plt.show()
```

```python
def cross_val_f1(X, y, model, model_name):
    
    kf = KFold(n_splits=2) #fazer com 30
    metric = make_scorer(f1_score, average='macro')

    scores = cross_val_score(model, X, y, cv=kf, scoring=metric)
    
    return pd.DataFrame( {'Model Name': model_name,
                          'F1 SCORE  CV': np.round( scores.mean(), 5 ).astype( str ) + ' +/- ' + np.round( scores.std(), 2 ).astype( str ) }, index=[0] )
```
